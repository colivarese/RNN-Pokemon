```python
import pandas as pd
```

```python
df = pd.read_csv('cleaned_data.csv', index_col=0)
```

```python
df.head()
```

```python
reviews = df.unstack().values
```

```python
corpus = dict()
for review in reviews:
    for char in review.split(' '):
        if char not in corpus:
            
            corpus[char] = 1
        else:
            corpus[char] += 1
```

```python
corpus
```

```python
sorted_corpus = {k: v for k, v in sorted(corpus.items(), key=lambda item: item[1], reverse=True)}
```

```python
cleaned_corpus = { k:v for k,v in corpus.items() if len(list(k)) > 2}
```

```python
cleaned_corpus
```

```python
def more_than_two_chars(row):
    words = row.split(' ')
    str = ''
    for word in words:
        if len(word) > 2:
            str += word
            str += ' '
    return str
```

```python
more_than_two_chars('Hi my name is Cesar')
```

```python
more_2_df = df['review'].apply(more_than_two_chars).to_frame()
```
